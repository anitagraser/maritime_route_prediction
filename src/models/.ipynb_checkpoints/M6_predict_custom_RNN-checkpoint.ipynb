```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edge(0, 1)
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
nx.draw_networkx(G)
```

```python
import torch
from torch import nn, optim
import torch.nn.functional as F
```

```python
# inputs
sequences = [[1,2,3], [2,3,4], [1,3,4]]
inputs = [sequence[:-1] for sequence in sequences]
gold_outputs = [sequence[1:] for sequence in sequences]
inputs_t = torch.tensor(inputs)
gold_outputs_t = torch.tensor(gold_outputs)

vocab_size = G.number_of_nodes()
embedding_size = 32
batch_size = len(sequences)
seq_len = len(gold_outputs[0])

# define network
# create an embedding of the input
embedding = nn.Embedding(vocab_size, embedding_size)
# rnn layer
rnn = nn.LSTM(embedding_size, embedding_size, batch_first=True)
# linear layer
e2v = nn.Linear(embedding_size, vocab_size)

# optimizer
opt = optim.Adam(lr=0.02, params = list(embedding.parameters()) + list(rnn.parameters()) + list(e2v.parameters()))



for epoch in range(100):
    # embedding layer
    x = embedding(inputs_t)
    # rnn layer
    emb_out, (h,c) = rnn(x)
    # Linear layer
    outputs = e2v(emb_out)

    # flatten outputs
    outputs_flat = outputs.view(batch_size*seq_len, vocab_size)
    gold_outputs_flat = gold_outputs_t.view(batch_size*seq_len)

    # loss function
    loss = F.cross_entropy(outputs_flat, gold_outputs_flat)
    # backprop
    opt.zero_grad()
    loss.backward()
    opt.step()
    print('loss %.4f' % loss)
    _,preds = outputs.max(dim=-1)
    
print('preds.size()', preds.size())
print('preds', preds)

```
